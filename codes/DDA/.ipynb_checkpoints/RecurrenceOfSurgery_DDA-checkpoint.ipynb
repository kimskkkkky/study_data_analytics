### Quest
 - apply() 이용
 - 대상 철럼 : 연령 --> 연령(중앙값)차이
 - 중앙값 - 개인 연령 = cell 작성
 

```python
age_mean = df_ROS_forpatient['연령'].mean()
```

```python
def mean_age(age) :
    result = age_mean - age
    return result
```

```python
df_ROS_forpatient['연령'].apply(mean_age)
```

```python
df_ROS_forpatient['연령(중앙값)'] = df_ROS_forpatient['연령'].apply(mean_age)
```
