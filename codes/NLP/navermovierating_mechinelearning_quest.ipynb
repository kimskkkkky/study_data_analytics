```python
import pandas as pd
df_naver_movie_comments = pd.read_csv('./ratings_train.txt', sep ='\t')
df_naver_movie_comments
```

```python
df_naver_movie_comments.columns
```

```python
df_naver_movie_comments['label'].value_counts()
```

```python
df_naver_movie_comments['label'] = df_naver_movie_comments['label'].replace({0: '부정', 1: '긍정'})
df_naver_movie_comments
```

```python
grouped_data = df_naver_movie_comments.groupby('label')
```

```python
comments_positive = grouped_data.get_group('긍정')
comments_negative = grouped_data.get_group('부정')
```

```python
len(comments_positive), len(comments_negative)
```

```python
comments_positive= comments_positive[:500]
comments_negative=comments_negative[:500]
```

```python
comments_positive = list(comments_positive['document'])
comments_positive
```

```python
comments_negative = list(comments_negative['document'])
comments_negative
```

```python
comments_list =comments_positive + comments_negative
```

```python
from mecab import MeCab
mecab = MeCab()

tokenized_comments = list()
for comment in comments_list :
    comment_morphs = mecab.morphs(comment)
    # print(' '.join(comment_morphs))
    tokenized_comments.append(comment_morphs)
tokenized_comments
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
comments = list()
for token in tokenized_comments :
    comments.append(' '.join(token))
```

```python
countVectorizer = CountVectorizer()
countVectorizer.fit(comments)
```

```python
countVectorizer.vocabulary_
```

```python
tfidfVectorizer = TfidfVectorizer(use_idf=True)
tfidfVectorizer.fit(comments)
```

```python
tfidfVectorizer.idf_
```

```python
feature = tfidfVectorizer.transform([comments[9]])  # 학습시 2차원을 이용했으므로 변환도 2차원 적용
feature.toarray() 
```

```python
features =tfidfVectorizer.transform(comments)
```

```python
target = ['긍정'] * len(comments_positive) + ['부정'] * len(comments_negative)
target
```

### 머신러닝 학습

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(features.toarray(), target)
```
