https://seaborn.pydata.org/generated/seaborn.barplot.html

estimator : string or callable that maps vector -> scalar, optional
Statistical function to estimate within each categorical bin.

```python
df_TOCC_sum = pd.pivot_table(data=df_TOCC, index='channel', values='amount', aggfunc=sum)
# index = 그룹바이, ascending=내림차순, iloc = index별로 원하는 갯수로 자름
df_TOCC_bottom_list = df_TOCC_sum.sort_values(by='amount', ascending=False).iloc[-7:].index.to_list()
df_TOCC_bottom_list
```

```python
sns.barplot(data=df_TOCC, x='channel', y='amount', estimator=sum, order=df_TOCC_bottom_list) # 채널별 하위 7개 순위 표현.
```

```python
sns.barplot(data=df_TOCC, x='channel', y='amount', estimator=sum, order=df_TOCC_bottom_list, hue = 'type_of_contract') # 채널별 하위 7개 순위 표현 및 
```

```python
df_TOCC_top_list = df_TOCC_sum.sort_values(by='amount', ascending=False).iloc[:3].index.to_list()
df_TOCC_top_list
```

```python
sns.boxplot(data=df_TOCC, x='channel', y='amount')
```

```python
sns.boxplot(data=df_TOCC, x='channel', y='amount', order = df_TOCC_top_list)
```
